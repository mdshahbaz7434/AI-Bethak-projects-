```python
def english(app_name):
    for i in app_name:
        if ord(i) < 128:
            return True
        else:
            return False
```

```python
def english(app_name):
    non_eng = 0 
    for i in app_name:
        if ord(i) > 127:
            non_eng +=1
    if non_eng >3:
        return False 
    else:
        return True 
            
```

```python
english("facebook")
```

```python
english("♥💖♥♥Facebook")
```
